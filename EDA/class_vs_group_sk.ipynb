```python
import pandas as pd
df_scores = pd.read_excel(r'../analysed_data/ROW_DATA_Sonja_add_MV_ECMO_and_ECMO_LPS.xlsx',sheet_name='Total Score',engine='openpyxl')
df_scores_Ct_MV_LP = df_scores[0:19]

df_scores
```

```python
df_Score_new = df_scores.iloc[:,[1,2,10]]
df_Score_new

Each_class_number_slide = {'Control':[0,0,0],'ECMO':[0,0,0],'ECMO+LPS':[0,0,0],'MV':[0,0,0],'MV+LPS':[0,0,0]}
for gr in range(0,len(df_Score_new['Group '])):
    score = df_Score_new['Median'][gr]
    tr_gr =  df_Score_new['Group '][gr]
    if score > 25:
        Each_class_number_slide[tr_gr][2]+=1
    elif score > 10:
        Each_class_number_slide[tr_gr][1]+=1
    else:
        Each_class_number_slide[tr_gr][0]+=1
Each_class_number_slide
```
