```python
RESULT_PATH = '../../../results/classification/predict_result_3cv/'
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pandas
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
%matplotlib inline


import re

import pandas as pd
import numpy as np
sns.set(font_scale=0.8)


model_names = ['EffnetB4_raw','Vgg16_raw','EffnetB4_man_aug',
               'Vgg16_rnd_aug','EffnetB4_rnd_aug','top_3_best_model_mv','top_3_best_model_MVote_slide','VGG16_binary_diff_th_3_slide']


folds = ['1st_part','2nd_part','3rd_part']
Score_dict = {0:'Low',1:'Medium',2:'High'}
df_models  = pd.DataFrame(data=None,columns =[model_names])

for model_name in model_names:
    keys = []
    Slide_dict = {}
    Slide_dict_True = {}
    Slide_dict_Predicted = {}
    for val_dir in folds:

        keys = []
        Slide_dict = {} 
        y_pred = []  
        with open(RESULT_PATH+model_name+'_'+val_dir+'_validation_predicted_newlabels.txt','r') as f:
            for line in f:
                y_pred.append(int(line.split(' ')[1].strip()))  

       ##############################################################################################    

        with open(RESULT_PATH+model_name+'_'+val_dir+'_validation_predicted_newlabels.txt','r') as fout: 
            for line in fout:
                S_num = str(line.split(' ')[0][0:8])
                la = int(line.split('_score_')[1].split('_')[0])
                if '.' in S_num:
                    S_num = S_num.replace('.','')
                if '_' in S_num:
                    S_num = S_num.replace('_','')
                if S_num not in keys:
                    keys.append(S_num)
                    Slide_dict[S_num]=[]
                    Slide_dict_True[S_num] = []
                    Slide_dict_Predicted[S_num] = []
                    if la > 25:
                        Score_real = 2
                    elif la > 10:
                        Score_real = 1
                    else:
                         Score_real= 0
                    Slide_dict_True[S_num]=Score_real
                
        #print(keys)
        fout.close()
        #########################################################
        ### Add all scores of tiles to the slide key
        with open(RESULT_PATH+model_name+'_'+val_dir+'_validation_predicted_newlabels.txt','r') as fout: 
            for line in fout:
                key_temp=line.split(' ')[0][0:8]
                if '.' in key_temp:
                    key_temp = key_temp.replace('.','')
                if '_' in key_temp:
                    key_temp = key_temp.replace('_','')

                Slide_dict[key_temp].append(float(line.split(' ')[1].strip()))

        #### Change all tile scores to the score of majority of tiles 
        y_pred_slide = np.ones(len(y_pred))  
        offset = 0
        for k in Slide_dict.keys():
            class_0 = [index for index, element in enumerate(Slide_dict[k]) if element == 0]
            class_1 = [index for index, element in enumerate(Slide_dict[k]) if element == 1]
            class_2 = [index for index, element in enumerate(Slide_dict[k]) if element == 2]
            majority_vote = np.argmax([len(class_0),len(class_1),len(class_2)])

            y_pred_slide[offset:offset+len(Slide_dict[k])]= majority_vote
            #print(y_pred_slide[offset:offset+len(Slide_dict[k])])
            offset += len(Slide_dict[k])
            Slide_dict_Predicted[k]=majority_vote
    #print(Slide_dict_Predicted.values())
    if 'Slide_number' not in df_models:
        df_models.insert(0, 'Slide_number', Slide_dict_True.keys()) 
    if 'Median_Score' not in df_models:
        df_models.insert(1, 'Median_Score', Slide_dict_True.values()) 
    #df_models['True_Score']= Slide_dict_True.values()
    df_models[model_name]= Slide_dict_Predicted.values()
    


df_models

df_models['Slide_number'] = df_models['Slide_number'].replace(['Slide28','slide28'],['Slide28_MV','slide28_Control'])


df_models



df_models.insert(1, 'Expert_2',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(1, 'Expert_1',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(1, 'Novice_1',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(1, 'Intermediate_2',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(1, 'Intermediate_1',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(6, 'Average_Score',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(7, 'pE1_pE2_Average',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(8, 'pE1_pE2_pE3_pE4_Median',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(1, 'Slide',np.zeros(len(df_models['Slide_number'].values)))
df_models.insert(1, 'Group',np.zeros(len(df_models['Slide_number'].values)))




df_models    
df_new_Score = df_scores.iloc[:,[1,2,3,4,5,6,7,11,15,16]]
df_new_Score


for i in range(len(df_models['Slide_number'].values)):
    k = df_models['Slide_number'].values[i]
    v = re.findall(r'\d+', str(k[0]))[0]
    
    if int(v) in df_new_Score['(sanned slide) Sample Name '].values:
        if int(v)!=28:
            scores = df_new_Score[df_new_Score['(sanned slide) Sample Name ']==int(v)].values[0][2:]
            group     = df_new_Score[df_new_Score['(sanned slide) Sample Name ']==int(v)].values[0][1]
            for l in range(len(scores)):
                if scores[l]>25:
                    scores[l] = 2
                elif scores[l]>10:
                    scores[l]= 1
                else:
                    scores[l] = 0
                    
            scores = np.insert(scores, 0, int(v))
            df_models.iloc[i,2:11] = scores
            df_models.iloc[i,1] = group
        elif int(v)==28:
            slide_group = str(k[0]).split('_')[1] 
            group = (df_new_Score[df_new_Score['(sanned slide) Sample Name ']==int(v)].values[:,1])
            x = np.where(group==slide_group)
            scores = df_new_Score[df_new_Score['(sanned slide) Sample Name ']==int(v)].values[x[0][0]][2:]
            for l in range(len(scores)):
                if scores[l]>25:
                    scores[l] = 2
                elif scores[l]>10:
                    scores[l]= 1
                else:
                    scores[l] = 0    
            scores = np.insert(scores, 0, int(v))
            df_models.iloc[i,2:11] = scores
            df_models.iloc[i,1] = slide_group
      
df_models  


df_models.iloc[29,2] = 28
df_models_plot = df_models.iloc[:,1:]
df_models_plot
df_models_plot.sort_values(by=('Group',), axis=0, ascending=True, inplace=True)
df_models_plot


df_models_plot.insert(2, "Novice_1", df_models_plot.pop("Novice_1"))
df_models_plot

df_models_plot = df_models_plot.rename(columns={"Novice_1": "N1", "Intermediate_1": "M1","Intermediate_2": "M2","Expert_1": "pE1","Expert_2": "pE2","Average_Score": "Average","Median_Score": "Median",
                       "EffnetB4_raw": "EN_r","Vgg16_raw": "V_r","EffnetB4_man_aug": "EN_m_a","Vgg16_rnd_aug": "V_r_a","EffnetB4_rnd_aug": "EN_r_a","top_3_best_model_mv": "T_3_t",
                       "top_3_best_model_MVote_slide": "T_t_s",'VGG16_binary_diff_th_3_slide':"V_bin_3"})
df_models_plot




fig, ax = plt.subplots(figsize=(15,6))         # Sample figsize in inches

 # labels for x-axis
y_axis_labels = df_models_plot.iloc[:,1] 



cols = ['Slide']

df_models_plot['Slide'] = df_models_plot[cols].apply(lambda x: int(x.values), axis=1)

cols = ['Group', 'Slide',]
df_label = pd.DataFrame()
df_label['combined'] = df_models_plot[cols].apply(lambda x: ' '.join(x.values.astype(str)), axis=1)


#s = sns.heatmap(,cmap = 'viridis', yticklabels=y_axis_labels,linewidths=.5, ax = ax)

#s.set(xlabel='Model', ylabel='Slide')
#s.set_xlabel('Model', fontsize=20)
#s.set_ylabel('Slide', fontsize=20)



#myColors = ((0.4, 0.7, 0.0, 1.0), (0.9, 0.8, 0.3, 1.0), (0.7, 0.0, 0.0, 1.0))
#cmap = LinearSegmentedColormap.from_list('Custom', myColors, len(myColors))

ax = sns.heatmap(df_models_plot.iloc[:,2:].transpose(), cmap='gray_r', linewidths=2,xticklabels=df_label['combined'] , linecolor='black',cbar=None)

# Manually specify colorbar labelling after it's been generated
#colorbar = ax.collections[0].colorbar
#colorbar.set_ticks([0.33, 1, 1.66])
#colorbar.set_ticklabels(['Low', 'Medium', 'High'],size=12,rotation=0)


# X - Y axis labels
ax.set_xlabel('Groups',fontsize=35)
ax.set_ylabel('',fontsize=35)

# Only y-axis labels need their rotation set, x-axis labels already have a rotation of 0
_, labels = plt.yticks()
_, labels_x = plt.xticks()
plt.setp(labels, rotation=0,fontsize=20)
plt.setp(labels_x, rotation=90,fontsize=20)



plt.show()
```

The following cell is for calculating the F1-score of individual scorers comapred to gold score which is median score

```python
import numpy as np
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report


df_scores  ## Have this from first column

df_scorer = df_scores.iloc[:,[1,2,3,4,5,6,7,10]]  ## We took only scorer columns as well as median 

df_scorer.insert(2, "Sinem novice ", df_scorer.pop("Sinem novice "))    ## Pop and push novice column to be first column with order of N1, M1, M2, pE1 and pE2

df_scorer = df_scorer.rename(columns={"Sinem novice ": "N1", "Sandra Intermediate": "M1","Martin Intermediate": "M2","Oscar Expert ": "pE1","Deniz Expert ": "pE2"})  ## rename previous column names to new ones

## This function is for classifying scores to 0, 1, and 2 or low, medium, and high damage classes

def classify(c):
   
    if (c > 25):
        return 2
    elif (c > 10):
        return 1
    else:
        return 0

df = df_scorer.iloc[:,2:8]

## The function will apply to all columns
df['N1'] = df['N1'].apply(lambda x: classify(x))
df['M1'] = df['M1'].apply(lambda x: classify(x))
df['M2'] = df['M2'].apply(lambda x: classify(x))
df['pE1'] = df['pE1'].apply(lambda x: classify(x))
df['pE2'] = df['pE2'].apply(lambda x: classify(x))
df['Median'] = df['Median'].apply(lambda x: classify(x))


## The f1score function
def fscore_func(col,median_col):
    return classification_report(col,median_col)


print('N1')
print(classification_report(df['N1'],df['Median']))

print('M1')
print(classification_report(df['M1'],df['Median']))

print('M2')
print(classification_report(df['M2'],df['Median']))

print('pE1')
print(classification_report(df['pE1'],df['Median']))

print('pE2')
print(classification_report(df['pE2'],df['Median']))
```
